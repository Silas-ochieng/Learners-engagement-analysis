# **Performing forward fill for the columns which had invalid input**
---

```python
# Specify the columns you want to check for invalid dates
date_columns = [
    "Learner SignUp DateTime", "Opportunity End Date", "Date of Birth",
    "Entry created at", "Apply Date",
]
# Function to check for invalid date formats
def find_invalid_dates(df, date_columns):
    invalid_entries = {}
    for column in date_columns:
        if column in df.columns:
            # Check for invalid dates
            invalid_mask = pd.to_datetime(df[column], errors='coerce').isna()
            invalid_entries[column] = df[invalid_mask]
    return invalid_entries

# Check for invalid dates
invalid_dates = find_invalid_dates(df, date_columns)

# Print results
for column, entries in invalid_dates.items():
    if not entries.empty:
        print(f"Invalid entries in '{column}':")
        print(entries)
    else:
        print(f"No invalid entries in '{column}'.")
```

# **Convert all date columns to datetime format**
---

```python
date_cols = [
    "Learner SignUp DateTime", "Opportunity End Date", "Date of Birth",
    "Entry created at", "Apply Date", "Opportunity Start Date"
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')  # invalid formats will become NaT
df.info()
```

```python
df.isnull().sum()
```

```python
# Perform forward fill for the specified date columns
date_columns = [
     "Learner SignUp DateTime", "Opportunity End Date", "Date of Birth",
    "Entry created at", "Apply Date",
]

# Fill forward for each specified column
for column in date_columns:
    df[column] = df[column].fillna(method='ffill')

# Verify the changes
print(df[date_columns].isnull().sum())  # Check for any remaining NaN values
```

# **Fill the two missing rows with unknown**
---

```python
df["Institution Name"].fillna("Unknown", inplace=True)
df["Current/Intended Major"].fillna("Undeclared", inplace=True)
```

# **Handling opportunity start date**
---
- Check the state of missing inputs

```python
missing_start = df[df["Opportunity Start Date"].isna()]
print(missing_start["Status Description"].value_counts())
```

- Rejected / Waitlisted / Applied
- These users never actually started the opportunity.
- Withdraw
- These might be learners who applied, got accepted, but chose to back out before or after a start date
- Drop out
- Drop out before attending the first session.

- Drop out early enough that no start date was ever logged (e.g., admin marked as drop out before onboarding completed).
This can occur because
-Registered but never showed up.

- Was marked as drop-out due to inactivity or failure to confirm attendance.



```python
# Create engagement target (1 = started, 0 = not started)
df_engaged = df[df["Opportunity Start Date"].notna()].copy()
df["Engaged"] = df["Opportunity Start Date"].notna().astype(int)
# Reset index
df.reset_index(drop=True, inplace=True)
df_engaged.reset_index(drop=True, inplace=True)
```

# **Normalize categorical values**
---

```python
df["Gender"] = df["Gender"].str.strip().str.title()  # e.g., "female" → "Female"
df["Country"] = df["Country"].str.strip().str.title()
df["Institution Name"] = df["Institution Name"].str.title()
df["Current/Intended Major"] =df["Current/Intended Major"].str.title()
df["Status Description"] = df["Status Description"].str.strip().str.title()
df["Opportunity Category"] =df["Opportunity Category"].str.strip().str.title()
df["Opportunity Name"] =df["Opportunity Name"].str.strip().str.title()
```

# **Feature Engineering**
 ---

```python
# Calculate age
df["Age"] = pd.to_datetime("today").year - df["Date of Birth"].dt.year

# Calculate time to apply
df["Apply Delay (days)"] = (df["Apply Date"] - df["Learner SignUp DateTime"]).dt.days

# Time between apply and opportunity start
df["Start Delay (days)"] = (df["Opportunity Start Date"] - df["Apply Date"]).dt.days

# Label drop-offs: started but never allocated (and past end date)
today = pd.to_datetime("today")
df["DropOff"] = df.apply(
    lambda x: 1 if (x["Status Description"] == "Started" and today > x["Opportunity End Date"]) else 0,
    axis=1
)

# --- Final preview ---
print("\nCleaned shape:", df.shape)
print("\nSample data:\n", df.head())
```

```python
# Save both versions
df.to_csv("cleaned_full_dataset.csv", index=False)
df_engaged.to_csv("engaged_users_only.csv", index=False)
```

```python
import matplotlib.pyplot as plt

# Map labels
engaged_labels = {1: 'Engaged (Started)', 0: 'Not Engaged (Drop-off)'}
engaged_counts = df["Engaged"].value_counts().rename(index=engaged_labels)

# Plot
colors = ['#4CAF50', '#F44336']  # Green for engaged, red for drop-off
ax = engaged_counts.plot(kind="bar", color=colors)

# Title and labels
ax.set_title("Engagement Distribution")
ax.set_xlabel("Engagement Status")
ax.set_ylabel("Number of Learners")

# Add percentage on top of bars
total = engaged_counts.sum()
for i, value in enumerate(engaged_counts):
    percent = (value / total) * 100
    ax.text(i, value + 10, f'{percent:.1f}%', ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate engagement rate by status
engagement_rate_by_status = df.groupby("Status Description")["Engaged"].mean().sort_values()

# Plot
colors = plt.cm.Set2.colors  # A pleasant color palette
ax = engagement_rate_by_status.plot(
    kind="bar",
    color=colors,
    figsize=(8, 5),
    title="Engagement Rate by Status"
)

# Axis labels
ax.set_ylabel("Engagement Rate")
ax.set_xlabel("Status Description")

# Add percentage labels on top of bars
for i, value in enumerate(engagement_rate_by_status):
    ax.text(i, value + 0.02, f"{value*100:.1f}%", ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Compute engagement rate
country_engagement = (
    df.groupby("Country")["Engaged"]
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)

# Plotly horizontal bar plot
fig = px.bar(
    country_engagement,
    x="Engaged",
    y="Country",
    orientation="h",
    text=country_engagement["Engaged"].apply(lambda x: f"{x*100:.1f}%"),
    color="Engaged",
    color_continuous_scale="Blues",
    title="Engagement Rate by Country",
    labels={"Engaged": "Engagement Rate"}
)

# Update trace to position text outside the bar
fig.update_traces(textposition="outside")

# Update layout to add gridlines
fig.update_layout(
    xaxis_tickformat=".0%",
    xaxis=dict(showgrid=True, gridcolor='lightgray', gridwidth=1),
    yaxis=dict(categoryorder="total ascending"),
    coloraxis_showscale=False,
    margin=dict(l=100, r=20, t=50, b=50)
)

fig.show()
```
